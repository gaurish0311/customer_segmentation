```python
x=df.iloc[:,[3,4]].values
```

```python
wcss = []
for i in range(1, 11):
    k = KMeans(n_clusters=i, init='k-means++', random_state=42)
    k.fit(x)
    wcss.append(k.inertia_)
```

```python
print(wcss)
```

```python
plt.plot(range(1,11),wcss)
plt.show()
```
